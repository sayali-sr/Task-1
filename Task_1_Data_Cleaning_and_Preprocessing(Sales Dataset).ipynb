### Import Necessary Libraries

```python
import pandas as pd
```

### Load the Dataset

```python
#Load the dataset
df = pd.read_csv('Sales Data.csv')
```

```python
#Print the dataset
df
```

### Preprocessing (Cleaning)

```python
#Missing values
df.isnull().sum()           # Check missing values
```

```python
#Drop unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df
```

```python
# Check and convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df
```

```python
# Remove rows with invalid dates (if any)
df = df[df['Order Date'].notnull()]
df
```

```python
#Original shape of the data
print(f"Original shape of the data:",df.shape)
```

```python
# Check for duplicates
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
```

```python
#Remove duplicates
df = df.drop_duplicates()
df
```

```python
print(f"New shape after removing duplicates: {df.shape}")
```

```python
# Drop rows with any missing values (optional: can also fillna)
df.dropna(inplace=True)
df
```

```python
# Now convert safely without warnings
df.loc[:, 'Quantity Ordered'] = df['Quantity Ordered'].astype(int)
df.loc[:, 'Price Each'] = df['Price Each'].astype(float)
df.loc[:, 'Sales'] = df['Sales'].astype(float)
df.loc[:, 'Month'] = df['Month'].astype(int)
df.loc[:, 'Hour'] = df['Hour'].astype(int)
```

```python
# Extract more time features
df.loc[:,'Day'] = df['Order Date'].dt.day
df.loc[:,'Day of Week'] = df['Order Date'].dt.day_name()
df['Year'] = df['Order Date'].dt.year
df
```

```python
#Top 10 most solf product
most_sold = df.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False).head(10)

print("Top 10 Most Sold Products:\n", most_sold)
```

```python
#Top 10 moset revenue generated product
most_revenue = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)

print("Top 10 Most Revenue-Generating Products:\n", most_revenue)
```

```python
# Save the cleaned dataset
df.to_csv("Cleaned_Sales_Data.csv", index=False)
```
